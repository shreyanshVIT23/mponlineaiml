```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
data = {"Age": [20, 25, 35], "Salary": [20000, 40000, 60000]}
df = pd.DataFrame(data)
df
```

```python
df["Annual Salary"] = df["Salary"]*12
df
```

```python
df["Age Group"] = pd.cut(
    df["Age"],
    bins=2,
    labels=["Young Adult", "Adult"]
)
df
```

```python
scaler = MinMaxScaler()
df["Salary Normalized"] = scaler.fit_transform(df[["Salary"]])
df
```
