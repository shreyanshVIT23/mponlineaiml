```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
data = {
    "Name": ["Amit", "Ravi", "Priya", "Neha"],
    "Age": [22, 30, 45, 35],
    "Salary": [25000, 50000, 75000, 60000],
    "Department": ["HR", "IT", "Finance", "HR"],
}
df = pd.DataFrame(data)
df
```

```python
df["AnnualSalary"] = df["Salary"] * 12
df
```

```python
df["AgeGroup"] = pd.cut(
    df["Age"], bins=[0, 25, 40, 100], labels=["Young", "Adult", "Senior"]
)
df
```

```python
df = pd.get_dummies(df, columns=["Department"])
df
```

```python
scaler = MinMaxScaler()
df["Salary"] = scaler.fit_transform(df[["Salary"]])
df
```
