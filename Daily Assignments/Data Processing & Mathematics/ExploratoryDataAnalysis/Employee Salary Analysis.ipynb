```python
import pandas as pd
import seaborn as sns
```

```python
data = {
    "Name": ["Amit", "Ravi", "Priya", "Neha", "Mohit"],
    "Age": [25, 32, 28, 35, 40],
    "Salary": [30000, 60000, 45000, 55000, 80000],
    "Department": ["HR", "IT", "IT", "HR", "Finance"],
}
df = pd.DataFrame(data)
df
```

```python
sns.histplot(df["Salary"])
```

```python
sns.boxplot(x=df["Salary"])
```

```python
sns.scatterplot(x="Age", y="Salary", data=df)
```

```python
sns.countplot(x="Department", data=df)
```

```python
df["Salary"].mean()
```

```python
average_salary = df["Salary"].mean()
print("1. Average Salary:", average_salary)
```

```python
df["Salary"].max()
```

```python
highest_salary = df["Salary"].max()
print("\n2. Highest Salary:", highest_salary)
```

```python
df.groupby("Department")["Salary"].mean().idxmax()
```

```python
dept_salary = df.groupby("Department")["Salary"].mean()

highest_salary_department = dept_salary.idxmax()

print("\n3. Department with Highest Average Salary:")
print(highest_salary_department)
```

```python
df["Age"].corr(df["Salary"])
```

```python
correlation = df["Age"].corr(df["Salary"])

print("\n4. Correlation Between Age and Salary:")
print(correlation)
```

```python
df["Department"].value_counts()
```

```python
department_distribution = df["Department"].value_counts()

print("\n5. Employees by Department:")
print(department_distribution)
```
