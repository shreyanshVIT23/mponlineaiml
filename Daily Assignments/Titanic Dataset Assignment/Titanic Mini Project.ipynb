```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv("./Titanic-Dataset.csv")
```

```python
df.info()
```

```python
df.head()
```

## Mean Median and Standard Deviation of Age

```python
mean_age = df["Age"].mean()
median_age = df["Age"].median()
std_age = df["Age"].std()

print("1. Age Statistics")
print("Mean Age:", mean_age)
print("Median Age:", median_age)
print("Standard Deviation:", std_age)
```

## Probablity Of Survival

```python
prob_survival = df["Survived"].mean()

print("\n2. Probability of Survival")
print("P(Survival) =", prob_survival)
```

### Conditional Probablity of Survival given gender is female

```python
female_passengers = df[df["Sex"] == "female"]

prob_survival_female = (
    female_passengers["Survived"].sum()
    / len(female_passengers)
)

print("\n3. Conditional Probability")
print("P(Survival | Female) =", prob_survival_female)
```

## Feature Matrix of X and Target Y

```python
X = df.drop(columns=["Survived", "PassengerId", "Name", "Ticket"])
y = df["Survived"]
print("\n4. Machine Learning Representation")
print("Feature Matrix Shape:", X.shape)
print("Target Vector Shape:", y.shape)
```

## Dummy Predictions, MAE & MSE

```python
dummy_predictions = np.ones(len(y))

mae = mean_absolute_error(y, dummy_predictions)
mse = mean_squared_error(y, dummy_predictions)

print("\n5. Model Evaluation")
print("MAE:", mae)
print("MSE:", mse)
```
